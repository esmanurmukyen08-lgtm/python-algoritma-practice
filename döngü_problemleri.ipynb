) Çift Sayıları Yazdırma (for)

Kullanıcıdan bir sayı al.

1'den o sayıya kadar olan çift sayıları yazdır.

Örnek:

Sayı: 10
2 4 6 8 10


```python
sayı=int(input("bir sayı giriniz:"))
for i in range(1,sayı+1):
  if i%2==0:
    print(i)
```

2) Toplam Hesaplama (while)

Kullanıcı sürekli sayı girsin.

0 girildiğinde giriş bitsin.

Girilen sayıların toplamını yazdır.

Örnek:

5
10
3
0

Toplam = 18


```python
toplam=0
while True:
  sayı=int(input("sayı giriniz:"))
  if sayı==0:
    print("sistem kapatılıyor")
    break
  else:
    toplam+=sayı

print("sayıların toplamı",toplam)
```

3) Şifre Kontrolü (while)

Doğru şifre:

python123

Kullanıcı doğru şifreyi girene kadar tekrar sorsun.

Doğru girince:

Giriş başarılı


```python
doğru_şifre="python123"
while True:
  şifre=input("şifrenizi giriniz:")
  if şifre==doğru_şifre:
    print("giriş başarılı")
    break
  else:
    print("yanlış şifre.tekrar deneyiniz")
```

4) Not Ortalaması Sistemi

Kullanıcı not girsin.

0 girerse sistem dursun.
Ortalama hesaplansın.
Kaç not girildiği yazılsın.


```python
toplam=0
sayaç=0
while True:
    puan=int(input("notunuzu giriniz:"))
    if puan==0:
        print("Sistem kapatılıyor.")
        break
    else:
        sayaç+=1
        toplam+=puan

if sayaç > 0:
    ortalama = toplam / sayaç
    print("Not ortalaması:", ortalama)
    print("Girilen not sayısı:", sayaç)
else:
    print("Hiç not girilmedi.")
```

5) Market Sepeti

Ürün fiyatları:

Ekmek = 10
Süt = 35
Su = 5

Kullanıcı ürün adı girsin.

"çıkış" yazana kadar devam etsin.

En sonunda toplam tutarı yazdır.



```python
toplam=0
ekmek=10
süt=35
su=5
while True:
  ürün=input("ürün adını giriniz:")
  if ürün=="çıkış":
    print("sistem kapatılıyor")
    break
  elif ürün=="ekmek":
    toplam+=10
  elif ürün=="süt":
    toplam+=35
  elif ürün=="su":
    toplam+=5
  else:
    print("hatalı giriş yaptınız")

print("toplam tutarınız:",toplam)
```

6) Personel Maaş Analizi

Kullanıcı maaşları girsin.

0 girildiğinde sistem bitsin.

Program:

Kaç personel girildi
En yüksek maaş
En düşük maaş

bilgilerini yazdırsın.


```python
sayaç=0
liste=[]
while True:
  maaş=int(input("personelin maaşını giriniz:"))
  if maaş==0:
    print("sistem kapatılıyor")
    break
  else:
    liste.append(maaş)
    sayaç+=1

print("personel sayısı:",sayaç)
if liste:
  print("en yüksek maaş",max(liste))
  print("en düşük maaş",min(liste))
else:
  print("Hiç maaş girilmedi.")
```

) Kargo Takip Sistemi

Kullanıcı paket ağırlığı girsin.

Kurallar:

0-5 kg   → 50 TL
5-10 kg  → 80 TL
10+ kg   → 120 TL

0 girilene kadar devam etsin.

Program sonunda:

Toplam kargo sayısı
Toplam gelir

yazılsın.


```python
kargo=0
gelir=0
while True:
  ağırlık=int(input("paket ağırlığını giriniz:"))
  if ağırlık==0:
   print("sistem kapatılıyor")
   break

  if 0<ağırlık<=5:
    kargo+=1
    gelir+=50

  elif 5<ağırlık<=10:
    kargo+=1
    gelir+=80

  elif ağırlık>10:
    kargo+=1
    gelir+=120

  else:
    print("hata verdi")

print("toplam kargo sayısı:",kargo)
print("toplam gelir:",gelir)
```

9) Çağrı Merkezi Şikayet Sistemi

Menü:

1 İnternet
2 Fatura
3 Nakil
4 Temsilci
0 Çıkış

Program sonunda:

İnternet şikayeti: X
Fatura şikayeti: X
Nakil şikayeti: X
Temsilci şikayeti: X
Toplam şikayet: X


```python
internet=0
fatura=0
nakil=0
temsil=0
while True:
  print("\n----çağrı merkezi şikayet sistemi-----")
  print("1-internet")
  print("2-fatura")
  print("3-nakil")
  print("4-temsilci")
  print("54-çıkış")

  seçim=input("bir seçim yapınız:")

  if seçim=="1":
    internet+=1
  elif seçim=="2":
    fatura+=1
  elif seçim=="3":
    nakil+=1
  elif seçim=="4":
    temsil+=1
  elif seçim=="5":
    print("sistem kapatılıyor")
    break

print("internet şikayeti:",internet)
print("fatura şikayeti",fatura)
print("nakil şikayeti",nakil)
print("temsilci şikayeti",temsil)
```
