11) Asal Sayı Kontrolü

Girilen sayının asal olup olmadığını kontrol et.

```python
sayı=int(input("sayıyı giriniz:"))

if sayı <= 1:
    print("asal sayı değildir")
elif sayı == 2:
    print("asal sayıdır")
else:
    is_prime = True
    for i in range(2, sayı):
        if sayı % i == 0:
            is_prime = False
            break
    if is_prime:
        print("asal sayıdır")
    else:
        print("asal sayı değildir")
```


13) Depo Stok Sistemi

Depoda ürünler:

laptop = 10 adet
mouse = 20 adet
klavye = 15 adet

Kullanıcı:

ürün adı
alınacak adet

girsin.

Kurallar:

stok yeterliyse satış yapılsın
stok güncellensin
yetersizse hata verilsin
“çıkış” yazınca sistem bitsin

Program sonunda kalan stokları yazdır.



```python
laptop = 10
mouse = 20
klavye = 15

while True:

    ürün = input("ürün adı (laptop/mouse/klavye/çıkış): ").lower()

    if ürün == "çıkış":
        print("Stok yönetim sisteminden çıkılıyor.")
        break

    if ürün not in ["laptop", "mouse", "klavye"]:
        print("Geçersiz ürün girdiniz. Lütfen 'laptop', 'mouse', 'klavye' veya 'çıkış' yazın.")
        continue

    try:
        adet = int(input("alınacak adet: "))
    except ValueError:
        print("Geçersiz adet. Lütfen bir sayı giriniz.")
        continue

    if adet <= 0:
        print("Geçersiz adet. Lütfen pozitif bir sayı giriniz.")
        continue

    # LAPTOP
    if ürün == "laptop":
        if adet <= laptop:
            laptop -= adet
            print("Satış başarılı.")
        else:
            print(f"Yetersiz stok. Mevcut laptop stoğu: {laptop} adet.")

    # MOUSE
    elif ürün == "mouse":
        if adet <= mouse:
            mouse -= adet
            print("Satış başarılı.")
        else:
            print(f"Yetersiz stok. Mevcut mouse stoğu: {mouse} adet.")

    # KLAVYE
    elif ürün == "klavye":
        if adet <= klavye:
            klavye -= adet
            print("Satış başarılı.")
        else:
            print(f"Yetersiz stok. Mevcut klavye stoğu: {klavye} adet.")

# SONUÇ
print("\n--- KALAN STOKLAR ---")
print(f"laptop: {laptop}")
print(f"mouse: {mouse}")
print(f"klavye: {klavye}")
```

14) ATM Sistemi

Başlangıç bakiyesi:

5000 TL

Kullanıcı:

para çekme
para yatırma
bakiye görüntüleme

işlemlerini yapabilsin.

Kurallar:

yetersiz bakiye olamaz
çıkış yapılana kadar sistem devam etsin


```python
başlangıç_bakiyesi=5000
while True:
  ekran=input("Yapmak istediğiniz işlemi seçiniz (çekme/yatırma/bakiye/çıkış): ").lower()

  if ekran=="çıkış":
    print("Ana menüye dönülüyor. İyi günler!")
    break # Exit the loop

  elif ekran=="çekme":
    çekilecek_para=int(input("Çekmek istediğiniz tutarı yazınız:"))
    if çekilecek_para <= 0:
      print("Geçersiz tutar. Pozitif bir sayı giriniz.")
    elif çekilecek_para > başlangıç_bakiyesi:
      print("Yetersiz bakiye.")
    else:
      başlangıç_bakiyesi -= çekilecek_para # Update the balance
      print(f"Para çekme işlemi başarılı. Yeni bakiyeniz: {başlangıç_bakiyesi} TL")

  elif ekran=="yatırma":
    yatırılacak_para=int(input("Yatırmak istediğiniz para tutarını giriniz:"))
    if yatırılacak_para <= 0:
      print("Geçersiz tutar. Pozitif bir sayı giriniz.")
    else:
      başlangıç_bakiyesi += yatırılacak_para # Update the balance
      print(f"Para yatırma işlemi başarılı. Yeni bakiyeniz: {başlangıç_bakiyesi} TL")

  elif ekran=="bakiye":
    print(f"Mevcut bakiyeniz: {başlangıç_bakiyesi} TL")
  else:
    print("Geçersiz işlem. Lütfen 'çekme', 'yatırma', 'bakiye' veya 'çıkış' yazınız.")
```
