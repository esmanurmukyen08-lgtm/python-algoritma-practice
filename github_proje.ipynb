1. Hepsi Pozitif mi?

Kullanıcıdan 5 sayı alın.
Eğer tüm sayılar pozitifse "Hepsi pozitif", değilse "Hepsi pozitif değil" yazdırın.

```python
liste=[]
pozitif_mi=True
for i in range(5):
  sayı=int(input("bir sayı giriniz:"))
  liste.append(sayı)
  if sayı<0:
    pozitif_mi=False

if pozitif_mi==True:
  print("hepsi pozitif")
else:
  print("hepsi pozitif değil")
```

5. Hepsi asal mı?

Kullanıcıdan 5 sayı alın.
Eğer tüm sayılar asal ise "Hepsi asal", değilse "Hepsi asal değil" yazdırın.

```python
hepsi_asal_mi = True
for i in range(5):
  sayı = int(input("bir sayı giriniz:"))

  # Numbers less than 2 are not prime
  if sayı < 2:
    hepsi_asal_mi = False
    continue # Skip further checks for this number, move to the next input

  # Assume the current number is prime until proven otherwise
  current_sayi_is_prime = True
  j = 2
  # Optimized primality test: check divisors up to the square root of 'sayı'
  while j * j <= sayı:
    if sayı % j == 0:
      current_sayi_is_prime = False # Found a divisor, so 'sayı' is not prime
      break # No need to check further divisors for this number
    j += 1

  # If the current number is not prime, then the overall condition "hepsi asal" is false
  if not current_sayi_is_prime:
    hepsi_asal_mi = False

# After checking all 5 numbers, print the final result
if hepsi_asal_mi:
  print("hepsi asal")
else:
  print("hepsi asal değil")
```

5. Şifre kontrolü

Kullanıcıya 3 hak verin.
Doğru şifre girilirse "Giriş başarılı", 3 yanlış denemede "Bloke edildi" yazdırın.

```python
doğru_şifre = 1234
hak = 3

while hak > 0:
  şifre = int(input("şifrenizi giriniz:")) # Convert input to integer
  if şifre == doğru_şifre:
    print("giriş başarılı")
    break # Exit loop if password is correct
  else:
    hak -= 1
    print(f"yanlış şifre,kalan hak: {hak}") # Fixed SyntaxError using f-string

# This check runs after the loop finishes
if hak == 0:
  print("bloke edildi")
```


7. 0 girene kadar devam

Kullanıcıdan sürekli sayı alın.
0 girildiğinde program dursun.

Program sonunda:

kaç tane pozitif sayı
kaç tane negatif sayı

olduğunu yazdırın.

```python
pozitif_sayı = 0
negatif_sayı = 0

while True: # Loop indefinitely until 0 is entered
  sayı = int(input("bir sayı giriniz:"))

  if sayı == 0:
    break # Exit the loop if 0 is entered

  if sayı > 0:
    pozitif_sayı += 1
  else:
    negatif_sayı += 1

print(f"Toplam pozitif sayı: {pozitif_sayı}")
print(f"Toplam negatif sayı: {negatif_sayı}")
```
