: Depo Sipariş Sistemi

Bir şirketin deposunda ürün stokları var:

laptop = 10
mouse = 25
klavye = 15

Kullanıcı sürekli ürün adı ve adet girsin.

Kurallar:

Ürün yoksa: "Böyle bir ürün yok"

Stok yeterliyse:

Satış başarılı
Kalan stok: ...

Stok yetmiyorsa:

Stok yetersiz

Kullanıcı ürün adına "çıkış" yazarsa program bitsin.

Program sonunda kalan tüm stokları yazdır.


```python
laptop=10
mouse=25
klavye=15
print("\n-----depo sipariş sistemi------")
while True:

  ürün=input("ürün adını giriniz:")

  if ürün=="çıkış":
    print("sistem kapatılıyor")
    break

  adet=int(input("ürün adetiniz giriniz:"))

  if ürün=="laptop":
    if adet<=laptop:
      laptop-=adet
      print("satış başarılı")
      print("kalan laptop:",laptop)
    else:
      print("stok yetersiz")

  elif ürün=="mouse":
    if adet<=mouse:
      mouse-=adet
      print("satıi başarılı")
      print("kalan stok:",mouse)
    else:
      print("stok yetersiz")

  elif ürün=="klavye":
    if adet<=klavye:
      klavye-=adet
      print("satış başarılı")
      print("kalan klavye:",klavye)
    else:
      print("stok başarısız")

else:
  print("böyle bir ürün bulunmamaktadır")

print("\n kalan stoklar")
print("kalan klavye:",klavye)
print("kalan mouse:",mouse)
print("kalan laptop:",laptop)




```

2) Algoritma Sorusu: Personel Maaş Zam Sistemi

Bir şirkette personelin maaşı giriliyor.

Kurallar:

Maaş 0 veya negatifse → Hatalı maaş
Maaş 0-17000 arasıysa → %30 zam
17000-30000 arasıysa → %20 zam
30000 üstüyse → %10 zam

Kullanıcı "çıkış" yazana kadar maaş girmeye devam etsin.


```python
while True:
  maaş=(input("personelin maaşını giriniz:"))

  if maaş=="çıkış":
    print("sistem kapatılıyor")
    break

  maaş=int(maaş)

  if maaş==0 or maaş<0:
    print("hatalı maaş")

  elif 0<maaş<=1700:
    yeni_maaş=maaş+(maaş*30/100)
    print(yeni_maaş)

  elif 1700<maaş<=3000:
    yeni_maaş=maaş+(maaş*20/100)
    print(yeni_maaş)

  elif maaş>3000:
    yeni_maaş=maaş+(maaş*10/100)
    print(yeni_maaş)

else:
  print("hatalı giriş")
```

3) Algoritma Sorusu: Çağrı Merkezi Şikayet Sistemi

Bir çağrı merkezinde müşteri şikayetleri alınıyor.

Kullanıcı şikayet türü girsin:

1 → İnternet arızası
2 → Fatura sorunu
3 → Nakil işlemi
4 → Müşteri temsilcisi

Yanlış seçimde hata versin.

Kullanıcı "0" girene kadar sistem devam etsin.

Program sonunda kaç tane şikayet alındığını yazdır.


```python
şikayet=0
while True:
  print("\n-----çağrı merkezi şikayet sistemi-----")
  print("1-internet arızası")
  print(
"2-fatura sorunu")
  print("3-nakil işlemi")
  print("4-müşteri temsilcisi")

  seçim_str = input("bir seçim yapınız:")

  if seçim_str == "0":
    print("Sistem kapatılıyor.")
    break

  try:
    seçim = int(seçim_str)
  except ValueError:
    print("Geçersiz giriş. Lütfen bir sayı giriniz (0-4 arası).")
    continue # Loop again for valid input

  if seçim == 1:
    şikayet+=1
    print("İnternet arızası şikayeti alındı.")
  elif seçim == 2:
    şikayet+=1
    print("Fatura sorunu şikayeti alındı.")
  elif seçim == 3:
    şikayet+=1
    print("Nakil işlemi şikayeti alındı.")
  elif seçim == 4:
    şikayet+=1
    print("Müşteri temsilcisi şikayeti alındı.")
  else:
    print("Hatalı seçim. Lütfen 1, 2, 3, 4 veya çıkış için 0 giriniz.")

print("Toplam alınan şikayet sayısı:", şikayet)
```

4) Algoritma Sorusu: Market Kasa İndirim Sistemi

Kullanıcı ürün fiyatı girsin.

Kurallar:

Fiyat negatifse → Hatalı giriş
0-500 TL → indirim yok
500-1000 TL → %10 indirim
1000 TL üstü → %20 indirim

Kullanıcı "0" girerse program bitsin.

Program sonunda toplam kaç TL satış yapıldığını yazdır.



```python
satış=0
while True:
  fiyat=int(input("ürün fiyatını giriniz:"))

  if fiyat==0:
    print("sistem kapatılıyor")
    break

  if fiyat<0:
    print("hatalı giriş")

  elif 0<fiyat<=500:
    print("indirim yok")
    satış+=fiyat
  elif 500<fiyat<=1000:
    indirim=fiyat-(fiyat*10/100)
    print(indirim)
    satış+=indirim
  elif fiyat>1000:
    indirim=fiyat-(fiyat*20/100)
    print(indirim)
    satış+=indirim

  else:
    print("hatalı giriş")

print("toplam satış:",satış)
```
