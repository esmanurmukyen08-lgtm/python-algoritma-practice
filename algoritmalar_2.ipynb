1) Mutlak Değer

Kullanıcıdan bir sayı al.

Negatifse pozitif yap
Pozitifse aynı yazdır

```python
sayı=int(input("bir sayı giriniz:"))
if sayı < 0:
  sayı = -sayı
print(sayı)
```

2) Yaş Kontrolü

Kullanıcı yaş girsin:

0–12 → Çocuk
13–17 → Genç
18–64 → Yetişkin
65+ → Yaşlı

```python
yaş=int(input("yaşınızı giriniz:"))
if 0<=yaş<=12:
  print("çocuk")
elif 13<=yaş<=17:
  print("genç")
elif 18<=yaş<=64:
  print("yetişkin")
elif yaş>=65:
  print("yaşlı")

else:
  print("geçersiz bir yaş giridniz")
```

3) Basamak Toplamı

Bir sayı gir:
Örn: 123 → 1+2+3 = 6


```python
sayı=int(input("sayı giriniz:"))
toplam=0
str_sayı=str(sayı)
for rakam in str_sayı:
  toplam+=int(rakam)

print(f"sayının toplamı={toplam}")
```

4) Ters Yazdırma

Girilen sayıyı ters çevir:

Örn:
1234 → 4321

```python
sayı=int(input("bir sayı giriniz:"))
str_sayı=str(sayı)
yeni_sayı=str_sayı[::-1]
print(yeni_sayı)
```

5) Çift Sayıları Sayma

1’den N’e kadar:

kaç tane çift sayı var?

Örn:
N=10 → 2,4,6,8,10 → 5 tane

```python
n=int(input("N sayısı giriniz:"))
toplam=0
for i in range(1,n+1):
  if i%2==0:
    toplam+=1
print(f"toplam {toplam} bu kadar çift sayı vardır")
```



Bir market var:

Ürünler:

ekmek = 7 TL
süt = 12 TL
yumurta = 3 TL
Sistem nasıl çalışacak?

Kullanıcı sürekli ürün seçer:

“ekmek”, “süt”, “yumurta”
“çıkış” yazınca sistem biter
Kurallar:
Her ürün sepete eklenir
Toplam fiyat hesaplanır
En sonda toplam ödeme yazdırılır
Bonus (hoca tuzağı):

Eğer kullanıcı yanlış ürün yazarsa:
👉 “Geçersiz ürün” yaz

```python
ekmek=7
süt=12
yumurta=3
toplam=0

while True:
  ürün=input("ürün seçisiniz(ekmek/süt/yumurta/çıkış):").lower()

  if ürün=="çıkış":
    break

  if ürün=="ekmek":
    toplam+=7
    print("ekmek sepete eklendi")

  elif ürün=="süt":
    toplam+=12
    print("süt sepete eklendi")

  elif ürün=="yumurta":
    toplam+=3
    print("sepete yumurta eklendi")

  else:
    print("geçersiz ürün")


print(f"toplsm fiyat={toplam} tl")
```
