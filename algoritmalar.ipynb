1) Tek/Çift Kontrolü

Soru:
Kullanıcıdan bir sayı al. Sayı çiftse "Çift", tekse "Tek" yazdır.

```python
sayı=int(input("bir sayı giriniz:"))
if sayı%2==0:
  print("çift sayıdır")
elif sayı%2!=0:
  print("tek sayıdır")
```

2) Not Sistemi

Soru:
Kullanıcıdan not al (0-100).

90-100 → A
80-89 → B
70-79 → C
50-69 → D
0-49 → F

```python
puan=int(input("notunuzu giriniz:"))
if 90<=puan<=100:
  print("notunuz A")
elif 80<=puan<=89:
  print("notunuz B")
elif 70<=puan<=79:
  print("notunuz C")
elif 50<=puan<=69:
  print("notunuz D")
elif 0<=puan<=49:
  print("notunuzF")
```

3) 1’den N’e Toplam

Soru:
Kullanıcı N girsin. 1’den N’e kadar olan sayıların toplamını hesapla.

```python
N = int(input("N giriniz:"))

toplam = 0

for i in range(1, N + 1):
    toplam += i

print(f"toplamı = {toplam}")
```

4) Çarpım Tablosu (mini)

Soru:
Kullanıcıdan bir sayı al. O sayının 1’den 10’a kadar çarpım tablosunu yazdır.

```python
sayı=int(input("bir sayı giriniz:"))
for i in range(1,11):
    print(f"{i}*{sayı}={i*sayı}")
```

5) Asal Sayı Kontrolü

Soru:
Girilen sayı asal mı değil mi kontrol et.

```python
sayı=int(input("bir sayı giriniz:"))
asal=True
for i in range(2,sayı):
  if sayı%i==0:
    asal=False
    break
else:
  print("asal sayıdır")
```

6) Şirket Stok Takip Sistemi (Final tipi 👇)

Bir şirketin 3 ürün stoğu var:

Laptop: 10 adet
Mouse: 25 adet
Klavye: 15 adet

Kullanıcıdan şu bilgiler alınacak:

Ürün adı (Laptop / Mouse / Klavye)
Satılan adet
Kurallar:
Eğer stok yeterliyse satış yapılır ve stok güncellenir
Eğer stok yetmiyorsa: "Yetersiz stok" yazdır
Kullanıcı "çıkış" yazana kadar sistem devam eder
Bonus:

En sonda toplam kaç ürün satıldığını da yazdır.

```python
laptop=10
mouse=25
klavye=15
toplam_satıs=0
while True:
  ürün_adı=input("ürün adını giriniz(laptop/mouse/klavye/çıkış):").lower()
  if ürün_adı=="çıkış":
    break

  adet=int(input("kaç adet satılacak"))

  if ürün_adı=="laptop":
    if laptop>=adet:
      laptop-=adet
      toplam_satıs+=adet
      print("satış başarılı")
    else:
      print("yetersiz stok")

  elif ürün_adı=="mouse":
    if mouse>=adet:
      mouse-=adet
      toplam_satıs+=adet
      print("satıs başarılı")
    else:
      print("yetersiz stok")

  elif ürün_adı=="klavye":
    if klavye>=adet:
      klavye-=adet
      toplam_satıs+=adet
      print("satıs başarılı")
    else:
      print("yetersiz stok")

  else:
    print("geçersiz ürün")

print(f"toplam satış={toplam_satıs}")
```
