```python
print("mini öğrenci not sistemi")
liste=[]
n=int(input("kaç ders var:"))
for i in range(n):
  ders=input("ders ismini giriniz:")
  vize=int(input("vize notu:"))
  final=int(input("final notu:"))
  ort=vize*0.4+final*0.6
  print(f" {ders} ortalaması : {ort}")

  harf = ""
  if ort>=90:
    print("AA")
    harf = "AA"
  elif 80<=ort<=89:
    print("BA/BB")
    harf = "BA/BB"
  elif 70<=ort<=79:
    print("CB")
    harf = "CB"
  elif 60<=ort<=69:
    print("CC")
    harf = "CC"
  elif 50<=ort<=59:
    print("DC")
    harf = "DC"
  elif ort<50:
    print("FF")
    harf = "FF"
  liste.append({"ders":ders, "ortalama":ort, "harf":harf})

if len(liste) > 0:
  ortalama_listesi = [item["ortalama"] for item in liste]
  genel_ortalama=sum(ortalama_listesi)/len(ortalama_listesi)
  if genel_ortalama>=60:
    print("geçti")
  else:
    print("kaldı")

print(f"{n} -----dersler----")
for d in liste:
  print(f"{d['ders']} |{d['ortalama']:.2f} | {d['harf']}")
```
