4-girilen kelimenin büyük ünlü uyumuna uyup uymadığını gösteren program yazınız

```python
kalın=["a","ı","o","u"]
ince=["e","i","ö","ü"]
while True:
  kalın_ünlü=False
  ince_ünlü=False
  metin=input("kelimeyi giriniz(programdan çıkmak için q):")
  if metin=="q":
    print("programdan çıkılıyo")
    break
  for harf in metin.lower(): # Kelimeyi küçük harfe çevirme eklendi
    if harf in kalın:
      kalın_ünlü=True
    if harf in ince:
      ince_ünlü=True

  if kalın_ünlü and ince_ünlü:
    print("büyük ünlü uyumuna uymaz")
  else:
    print("büyük ünlü uyumuna uyar")
```

 10-Basit ATM Sistemi (if-else + while)

Başlangıç bakiyesi: 1000 TL

Menü:

Para çek
Para yatır
Bakiye göster
Çıkış



```python
başlangıç_baakiyesi=1000
doğru_şifre=1234
hak=3

while hak>0:
  şifre=int(input("bir şifre giriniz:"))
  if şifre==doğru_şifre:
    print("şifreniz doğrudur ana menüye geçiniz")
    break # Exit the loop if password is correct
  else:
    hak-=1
    print(f"şifreniz yanlıştır. Kalan hakkınız: {hak}. Tekrar deneyiniz")

if hak == 0:
  print("Hesabınız bloke olmuştur. Lütfen bankanızla iletişime geçiniz.")
else:
  print("\n--- ATM Menüsü ---")
  print("1 = Para Çekme")
  print("2 = Para Yatırma")
  print("3 = Bakiye Görüntüleme")
  print("4 = Çıkış")

  seçim=int(input("Bir seçim yapınız (1-4): "))

  if seçim==1:
    çekilecek_miktar=int(input("Çekilecek miktarı yazınız:"))
    if başlangıç_baakiyesi>=çekilecek_miktar:
      yeni_bakiye=başlangıç_baakiyesi-çekilecek_miktar
      başlangıç_baakiyesi = yeni_bakiye # Update the balance
      print(f"Yeni bakiyeniz: {yeni_bakiye} TL")
    else:
      print("Bakiyeniz yetersiz.")

  elif seçim==2:
    yatırılacak_miktar=int(input("Yatırılacak miktarı giriniz:"))
    yeni_bakiye=başlangıç_baakiyesi+yatırılacak_miktar
    başlangıç_baakiyesi = yeni_bakiye # Update the balance
    print(f"Yeni bakiyeniz: {yeni_bakiye} TL")

  elif seçim==3:
    print(f"Bakiyeniz: {başlangıç_baakiyesi} TL")

  elif seçim==4:
    print("Program sonlanmıştır. İyi günler dileriz.")

  else:
    print("Geçersiz seçim. Lütfen 1 ile 4 arasında bir sayı giriniz.")
```

 11-Asal Sayı Kontrolü (for + if)

Kullanıcıdan sayı al.

Asal mı değil mi kontrol et

```python
while True:
    sayı=int(input("bir sayı giriniz(çıkış='0'):"))

    if sayı==0:
        print("programdan çıkılıyo")
        break

    asal_mi=True

    if sayı<2:
        asal_mi=False
    else:
        j=2
        while j<sayı:
            if sayı%j==0:
                asal_mi=False
                break
            j+=1

    if asal_mi:
        print("sayı asaldır")
    else:
        print("sayı asal değildir")
```

 13-Fibonacci Serisi (while veya for)

Kullanıcı kaç terim istiyorsa o kadar Fibonacci üret:.

```python
liste=[0,1]
n=int(input("kaç terim istediğinizi yazınız:"))

for i in range(2,n):
  yeni=liste[i-1]+liste[i-2]
  liste.append(yeni)
print(liste)
```
