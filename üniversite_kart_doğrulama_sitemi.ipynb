Üniversite Kart Doğrulama Sistemi

Üniversite öğrenci kartları 9 hanelidir.

Kurallar:

İlk 4 hane giriş yılını göstermektedir.
Sonraki 2 hane bölüm kodudur.
11 → YBS
22 → İşletme
33 → İktisat
Son 3 hane öğrenci sıra numarasıdır.
Hatalı kart numaralarında uyarı verilmelidir.
Kullanıcı "0" girene kadar sistem çalışmalıdır.
Program sonunda her bölümden kaç öğrenci giriş yaptığını yazdırınız.

```python
ybs=0
işletme=0
iktisat=0
while True:
  kart_no=input("kart numaranızı giriniz:")
  if kart_no=="0":
    print("sistem kapatılıyor.")
    break

  if len(kart_no)!=9:
    print("hatalı sayı girdiniz.kart numarası 9 olmalıdır")
    continue

  if not kart_no.isdigit():
    print("kart sadece rakamlardan oluşmaktır.")
    continue

  giriş_yılı=kart_no[0:4]
  bölüm_kod=kart_no[4:6]
  sıra_numarası=kart_no[6:9]

  if bölüm_kod=="11":
    print("ybs öğrencisi")
    ybs+=1

  elif bölüm_kod=="22":
    print("işletme öğrencisi")
    işletme+=1

  elif bölüm_kod=="33":
    print("iktisat öğrencisi")
    iktisat+=1

  else:
    print("geçersiz bölüm kodu")


print("-----program sonu-----")
print("ybs öğrenci sayısı:",ybs)
print("işletme öğrenci sayısı:",işletme)
print("iktisat öğrnci sayısı",iktisat )
```
